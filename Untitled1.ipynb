```python
# PDF se text extract karne ka code
from google.colab import files
import pdfplumber

# PDFs upload karo
uploaded = files.upload()

# Saara text ek file mein save karo
all_text = ""
for filename in uploaded:
    if filename.endswith(".pdf"):
        print(f"Extracting: {filename}")
        with pdfplumber.open(filename) as pdf:
            for page in pdf.pages:
                all_text += page.extract_text() + "\n"

# Text ko file mein save karo
with open("dsa_corpus.txt", "w", encoding="utf-8") as f:
    f.write(all_text)

print("Ho gaya! Saara text 'dsa_corpus.txt' mein save ho gaya.")
```

```python
!pip install sentence-transformers chromadb

from sentence_transformers import SentenceTransformer
import chromadb

# Text ko chunks mein divide karo
with open("dsa_corpus.txt", "r") as f:
    text = f.read()

chunks = [text[i:i+1000] for i in range(0, len(text), 1000)]

# AI ko text samjhao (Embeddings banayein)
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(chunks)

# ChromaDB mein store karo (Taaki baad mein search ho sake)
client = chromadb.Client()
collection = client.create_collection("dsa_knowledge")

for i, chunk in enumerate(chunks):
    collection.add(
        documents=[chunk],
        ids=[f"chunk_{i}"]
    )

print("AI ne DSA samajh liya! Ab sawal pooch sakte ho.")
```

```python
# First install LangChain if not already installed
!pip install langchain

# Now import the required class
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Then use the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,  # Larger chunks for better context
    chunk_overlap=200  # Overlap to maintain continuity
)
```

```python
from moviepy.editor import *
from moviepy.video.tools.subtitles import SubtitlesClip

def generate_video(query, output_file="output.mp4"):
    # Step 1: AI se detailed answer lo
    results = collection.query(query_texts=[query], n_results=2)  # 2 chunks = zyada detail
    answer = " ".join(results["documents"][0])

    # Step 2: Audio generate karo
    tts = gTTS(answer, lang="en")
    tts.save("audio.mp3")

    # Step 3: Background + Captions banayo
    bg = ColorClip((1280, 720), color=(0, 0, 100), duration=10)
    audio = AudioFileClip("audio.mp3")

    # Captions ke liye
    subtitles = []
    for i, sentence in enumerate(answer.split(". ")):
        txt_clip = TextClip(sentence, fontsize=30, color="white", bg_color="black")
        subtitles.append(txt_clip.set_start(i * 2).set_duration(2))

    # Video combine karo
    final_video = CompositeVideoClip([bg] + subtitles).set_audio(audio)
    final_video.write_videofile(output_file, fps=24)
```

```python
!pip install moviepy gtts pillow numpy transformers

from moviepy.editor import *
from gtts import gTTS
from PIL import Image, ImageDraw, ImageFont
import numpy as np
from transformers import pipeline

# ======================
# 1. USER PROMPT INPUT
# ======================
topic = input("Kon sa DSA topic samajhna chahte ho?\nExamples: Graphs, Binary Search, Linked List\nEnter topic: ")

# ======================
# 2. GENERATE EXPLANATION (Using AI)
# ======================
generator = pipeline('text-generation', model='gpt2')
prompt = f"Explain {topic} in Data Structures with examples in 100 words:"
explanation = generator(prompt, max_length=150)[0]['generated_text']

# ======================
# 3. CREATE VIDEO WITH ANIMATIONS
# ======================
def create_frame(text, bg_color=(0, 72, 186), text_color=(255, 255, 0)):
    img = Image.new('RGB', (800, 600), color=bg_color)
    d = ImageDraw.Draw(img)

    # Title
    d.text((50, 50), f"DSA: {topic}", fill=(255, 255, 255), font_size=50)

    # Bullet points
    for i, line in enumerate(text.split('. ')):
        d.text((80, 150 + i*40), f"• {line}", fill=text_color, font_size=30)

    return np.array(img)

# Generate frames with animations
clips = []
for i in range(0, len(explanation), 50):
    frame_text = explanation[i:i+50]
    frame = ImageClip(create_frame(frame_text)).set_duration(3)
    clips.append(frame.crossfadein(1))

# ======================
# 4. ADD AUDIO & CAPTIONS
# ======================
tts = gTTS(explanation, lang='en')
tts.save("audio.mp3")

final_video = concatenate_videoclips(clips)
final_video = final_video.set_audio(AudioFileClip("audio.mp3"))

# ======================
# 5. EXPORT VIDEO
# ======================
final_video.write_videofile(f"{topic}_explainer.mp4", fps=24, codec='libx264')

print(f"Video '{topic}_explainer.mp4' ready!")
from google.colab import files
files.download(f"{topic}_explainer.mp4")
```
